```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
colums=['user_id','item_id','rating','timestamp']
df=pd.read_csv("ml-100k/u.data",sep='\t',names=colums)
colums2=['item_id','title']
it=pd.read_csv("ml-100k/u.item",sep='\|',header=None)
it=it[[0,1]]
it.columns=colums2
df=pd.DataFrame(pd.merge(df,it,on='item_id'))
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
ratings['num_of_rating']=df.groupby('title').count()['rating']
movie_mat=df.pivot_table(index="user_id",columns='title',values='rating')
def pred(movie_name):
    movie_user_ratings=movie_mat[movie_name]
    similar=movie_mat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar,columns=['correlation'])
    corr_movie=corr_movie.dropna()
    corr_movie=pd.merge(corr_movie,ratings,on='title')
    predictions=corr_movie[corr_movie['num_of_rating']>100].sort_values('correlation',ascending=False)
    return predictions.head(n=10)
pred('Star Wars (1977)')
```

```python
import numpy as np
def simulateProblem():
    conf=np.random.randint(2)   ## if conf==0 means (a,2a) configuartion 
    if conf==0:                 ## esle if conf ==1  means (2a,a) 
        a,b=(1,2)
    else:
        a,b=(2,1)
    first=np.random.randint(2)
    ##now here generate four cases
    ##if first==0: conf==0 ##means you are on the (a,2a) configuration initially
    ## if first ==1 : conf==0 means you are on the (2a,a) configuration initially    
    ##if first==0: conf==1 ##means you are on the (2a,a) configuration initially
    ## if first ==1 : conf==1 means you are on the (a,2a) configuration initially    
    ## so we can use XOR operation to contract it since xor return 0 when number are same and 
    ## non zero when num are different
    ## if first^conf==0 means standing on (a,2a) config 
    ## if first^conf==1 means standing on (2a,a) config
    first_index=np.random.randint(2)
    switch_or_not=np.random.randint(2)
    final_index=(first_index+switch_or_not)%2  ## (first_index+switch_or_not)%2 is the final index stand
    won=0
    loose=0

    if first^conf==0:  ## means(a,2a) configuration
        if final_index==0:  ##we loose 
            loose=1
        elif final_index==1:  ##we won
            won=1
    elif first^conf==1:
        if final_index==0:  ##we won
            won=1
        elif final_index==1:  ##we loose
            loose=1
    a=False
    b=False
    
    if won==1:
        a=True
    if switch_or_not==0:
        b=True
    return (a,b)    
assert type(simulateProblem()[0])==bool
assert type(simulateProblem()[0])==bool


#######################################################
def run_simulation():
    win_due_to_sticking=0;
    win_due_to_switching=0
    total_iter=10000
    for i in range(total_iter):   ##running 10000 times
        a,b=simulateProblem()
        if a==True and b==True:
            win_due_to_switching+=1
        elif a==True and b==False:
            win_due_to_sticking+=1
    prob_w_stick=win_due_to_sticking/total_iter 
    print("win due to sticking are : "+str(win_due_to_sticking))
    print("win due to switching are : "+str(win_due_to_switching))
    prob_w_switch=win_due_to_switching/total_iter
    return (prob_w_stick,prob_w_switch)
######################################################################
p_win_due_to_sticking,p_win_due_to_switching=run_simulation()
print("probability of win due to sticking is: "+str(p_win_due_to_sticking))
print("probability of win due to switching is: "+str(p_win_due_to_switching))
```
