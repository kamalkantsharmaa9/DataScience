```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('mnist\mnist_test.csv')
x=df.values
split=int(0.8*x.shape[0])
x_train=x[0:split,1:]
y_train=x[0:split,0]
x_test=x[split:,1:]
y_test=x[split:,0]
def imgview(mat):
    plt.imshow(mat.reshape(28,28),cmap='gray')
    plt.show()
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
def knn(X,Y,querypoints,k=5):
    vars=[]
    m=X.shape[0]
    for i in range (m):
        d=dist(querypoints,X[i])
        vars.append([d,Y[i]])
    vars=sorted(vars)
    vars=vars[:k]
    vars=np.array(vars)
    newval=np.unique(vars[:,1],return_counts=True)
    index=newval[1].argmax()
    pred=newval[0][index]
    return pred
i=510
val=knn(x_train,y_train,x_test[i])
print(val)
print(y_test[i])
imgview(x_test[i])
```

```python
def accu(y_test,x_train,y_train,x_test):
    m=y_test.shape
    corr=0
    m=m[0]
    t=10
    for i in range(t):
        val=knn(x_train,y_train,x_test[i]) 
        print(val)
        print(y_test[i])
        if val == y_test[i]:
            corr=corr+1
    return ((corr*100)/(t))   
neww=accu(y_test,x_train,y_train,x_test)
print(neww)
```
